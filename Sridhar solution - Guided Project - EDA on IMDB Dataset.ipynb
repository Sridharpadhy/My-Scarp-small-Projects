# Converting string into proper list 

```python
#we will do this coversion on this 3 colum with the help of eval function wich does is like take any string and count it as phyton expersion.
#eg-
x=4
print(eval('x==5'))
# if u dont use eval fn it will just print the x==5 but now its act as a boolean function.
```

```python
# this eval fn and conversion is done only on non null values.with the help of the apply fn of panda
def convert_to_list  (str):
 return eval(str)
 #this fn is created so that we can use it in apply fn.
```

```python
imdb_df.loc[~imdb_df['genres'].isna(),'genres'] = df.loc[~df['genres'].isna(),'genres'].apply(convert_to_list)
#so we have used only for non null values and we have use df here because in imdf which is new and needed df and we have to store this strng to list data in it.
#and also if we use both side the imdb_df the the eval fn which is applied 1 take one strng and return to imdb_df and made it list now for 2nd the eval fn will 
# not take list it required strng only so we used df from which the strng will be keep on coming and get converted into list and stored into imdb_df.b 
```

```python
# same for cast 
imdb_df.loc[~imdb_df['cast'].isna(),'cast'] = df.loc[~df['cast'].isna(),'cast'].apply(convert_to_list)
```

```python
# same for crew 
imdb_df.loc[~imdb_df['crew'].isna(),'crew'] = df.loc[~df['crew'].isna(),'crew'].apply(convert_to_list)
```

# 9.Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI? 

```python
imdb_df.head() # this df was made in roi question
```

```python
no_nadf = imdb_df[~imdb_df['crew'].isna()]
```

```python
no_nadf['profit'] = round(no_nadf['revenue']-no_nadf['budget'])
no_nadf['roi'] = round((no_nadf['profit']/no_nadf['budget'])*100)
```

```python
no_nadf.shape
no_nadf.head()
```

```python
#A simple function to extract list of all producers for a given movie_index
def create_producer_list(index):
  crew_list= no_nadf.loc[index,'crew']
# A function is created which will takes a index and give crew list.
  producer_list=[]
  for elem in crew_list:
     if elem['job']=='Producer':
        producer_list.append(elem['name'])
  return producer_list
```

```python
create_producer_list (1) #whatever index u give it will give producer list
```

```python
#same for director.
def create_director(index):
  
  crew_list= no_nadf.loc[index,'crew']

  for elem in crew_list:
     if elem['job']=='Director':
        return elem['name']
```

```python
#now we are creating empty dataframe Table to append data in it .
Table = pd.DataFrame(columns=['Movie Title','Producers','Directors','ROI'])
```

```python
# now we will append data in it.
for index,row in no_nadf.iterrows():
  
  try:
      Table = Table.append({'Movie Title':no_nadf.loc[index,'original_title'],'Producers':create_producer_list(index),'Directors':create_director(index),'ROI':no_nadf.loc[index,'roi']}, ignore_index=True)
  except:
    continue
    
```

```python
Table
```

```python
#for the top 3 producer of average roi.
(Table.sort_values('ROI', ascending=False)).head(3)
```
